```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
X_train = np.linspace(0.,1.,201) # training set
X_test = np.linspace(0.,1.,101) # testing set

np.random.seed(8681)
t_train = np.sin(4*np.pi*X_train) + 0.3 * np.random.randn(201)
t_test = np.sin(4*np.pi*X_test) + 0.3 * np.random.randn(101)
t_actual = np.sin(4*np.pi*X_train) 
```

```python
shuffle_data = np.column_stack((X_train, t_train))
np.random.shuffle(shuffle_data)
plt.figure()
plt.scatter(shuffle_data[:,0],shuffle_data[:,1],color = 'blue',label="prediction Points")
plt.scatter(X_train,t_train,color = 'orange',label="Testing Points")
plt.plot(X_train,t_actual,color = "red",label="True curve")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  
```

```python
#implement k nearest neighbours k range from 1 to 60 inclusive
k = list(range(1,61))
print(k)
k = 1
prediction_points = []
def compute_distance(point_index,k):
  total_distance = []
  for i in range(len(X_train)):
    #Euclidian distance
    distance = math.sqrt(pow(X_train[i] - X_train[point_index],2)+pow(t_train[i] -t_train[point_index],2))
    
    total_distance.append([distance,t_train[i]])
  #sort the list in ascending order
  total_distance.sort(key=lambda x: x[0])
  # print(total_distance)
  # y_avg = sum(total_distance[1:k+1]) / k
  total_sum = 0
  for i in range(k):
    total_sum = total_sum + total_distance[i+1][1]
    
  return total_sum/k
    
for i in range(len(X_train)):
  # print(X_train[i])
  prediction_points.append(compute_distance(i,k=3))
  
print("the preidiction points are",prediction_points)

def plt_prediction():
#function to plot train and test points
  plt.figure(2)
  plt.scatter(X_train,prediction_points,color = 'orange',label="prediction Points")
  # plt.scatter(X_test,t_test,color = 'blue',label="Testing Points")
  plt.plot(X_train,t_actual,color = "red",label="True curve")
  plt.xlabel('x')
  plt.ylabel('y')
  plt.legend()  
  
def plt_origin():
  #function to plot train and test points
  plt.figure(1)
  plt.scatter(X_train,t_train,color = 'orange',label="Training Points")
  # plt.scatter(X_test,t_test,color = 'blue',label="Testing Points")
  plt.plot(X_train,t_actual,color = "red",label="True curve")
  plt.xlabel('x')
  plt.ylabel('y')
  plt.legend()  
```
